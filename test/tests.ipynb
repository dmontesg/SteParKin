### Test 1: Add a new group

In this test, we aggregate a new group. To do so, we provide a custom file for the parameters of the associations and moving groups and give a color (#ffff00 = yellow) for that group in the dict_colors_groups. We also obtain the Boettlinger diagrams independently.
